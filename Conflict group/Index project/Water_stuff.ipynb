```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Function to create figure for whatever region we want

def region_specific_figure(region):
    ssp1=pd.read_csv("wat_consum_by_region_ssp1.csv")
    ssp1_mun = ssp1[ssp1["input"] == "water_td_muni_C"]

    ssp2=pd.read_csv("wat_consum_by_region_ssp2.csv")
    ssp2_mun = ssp2[ssp2["input"] == "water_td_muni_C"]

    ssp3=pd.read_csv("wat_consum_by_region_ssp3.csv")
    ssp3_mun = ssp3[ssp3["input"] == "water_td_muni_C"]

    ssp4=pd.read_csv("wat_consum_by_region_ssp4.csv")
    ssp4_mun = ssp4[ssp4["input"] == "water_td_muni_C"]

    ssp5=pd.read_csv("wat_consum_by_region_ssp5.csv")
    ssp5_mun = ssp5[ssp5["input"] == "water_td_muni_C"]

    #filter region
    ssp1_mun_reg = ssp1_mun[ssp1_mun["region"] == region]
    ssp2_mun_reg = ssp2_mun[ssp2_mun["region"] == region]
    ssp3_mun_reg = ssp3_mun[ssp3_mun["region"] == region]
    ssp4_mun_reg = ssp4_mun[ssp4_mun["region"] == region]
    ssp5_mun_reg = ssp5_mun[ssp5_mun["region"] == region]

    sns.lineplot(data = ssp1_mun_reg, x = "year", y ="value", label = "SSP1 - Sustainability")
    sns.lineplot(data = ssp2_mun_reg, x = "year", y ="value", label = "SSP2 - Middle of the Road")
    sns.lineplot(data = ssp3_mun_reg, x = "year", y ="value", label = "SSP3 - Regional Rivalry")
    sns.lineplot(data = ssp4_mun_reg, x = "year", y ="value", label = "SSP4 - Inequality")
    sns.lineplot(data = ssp5_mun_reg, x = "year", y ="value", label = "SSP5 - Fossil Fuels")
    plt.title("Municipal Water Consumption across SSPs " + region)
    plt.xlabel("Year")
    plt.ylabel("Municipal Water Consumption (km3)")
    plt.legend()

region_specific_figure("USA")
```

```python
def make_consum_by_pop(region):
    ssp1=pd.read_csv("wat_consum_by_region_ssp1.csv")
    ssp1_mun = ssp1[ssp1["input"] == "water_td_muni_C"]
    ssp1_mun_reg = ssp1_mun[ssp1_mun["region"] == region]
    ssp1_mun_reg.reset_index(inplace = True)
    ssp1_pop = pd.read_csv("pop_by_region_ssp1.csv")
    ssp1_pop_reg = ssp1_pop[ssp1_pop["region"] == region]
    ssp1_pop_reg_1 = ssp1_pop_reg.rename(columns={'value': 'Pop(thousands)'})
    ssp1_pop_reg_1.reset_index(inplace = True)
    data1 = pd.concat([ssp1_mun_reg, ssp1_pop_reg_1["Pop(thousands)"]], axis = 1)
    data1["consum by pop (km3/thousand person)"] = data1["value"]/data1["Pop(thousands)"]
    
    ssp2=pd.read_csv("wat_consum_by_region_ssp2.csv")
    ssp2_mun = ssp2[ssp2["input"] == "water_td_muni_C"]
    ssp2_mun_reg = ssp2_mun[ssp2_mun["region"] == region]
    ssp2_mun_reg.reset_index(inplace = True)
    ssp2_pop = pd.read_csv("pop_by_region_ssp2.csv")
    ssp2_pop_reg = ssp2_pop[ssp2_pop["region"] == region]
    ssp2_pop_reg_1 = ssp1_pop_reg.rename(columns={'value': 'Pop(thousands)'})
    ssp2_pop_reg_1.reset_index(inplace = True)
    data2 = pd.concat([ssp2_mun_reg, ssp2_pop_reg_1["Pop(thousands)"]], axis = 1)
    data2["consum by pop (km3/thousand person)"] = data2["value"]/data2["Pop(thousands)"]

    ssp3=pd.read_csv("wat_consum_by_region_ssp3.csv")
    ssp3_mun = ssp3[ssp3["input"] == "water_td_muni_C"]
    ssp3_mun_reg = ssp3_mun[ssp3_mun["region"] == region]
    ssp3_mun_reg.reset_index(inplace = True)
    ssp3_pop = pd.read_csv("pop_by_region_ssp3.csv")
    ssp3_pop_reg = ssp3_pop[ssp3_pop["region"] == region]
    ssp3_pop_reg_1 = ssp3_pop_reg.rename(columns={'value': 'Pop(thousands)'})
    ssp3_pop_reg_1.reset_index(inplace = True)
    data3 = pd.concat([ssp3_mun_reg, ssp3_pop_reg_1["Pop(thousands)"]], axis = 1)
    data3["consum by pop (km3/thousand person)"] = data3["value"]/data3["Pop(thousands)"]

    ssp4=pd.read_csv("wat_consum_by_region_ssp4.csv")
    ssp4_mun = ssp4[ssp4["input"] == "water_td_muni_C"]
    ssp4_mun_reg = ssp4_mun[ssp4_mun["region"] == region]
    ssp4_mun_reg.reset_index(inplace = True)
    ssp4_pop = pd.read_csv("pop_by_region_ssp4.csv")
    ssp4_pop_reg = ssp4_pop[ssp4_pop["region"] == region]
    ssp4_pop_reg_1 = ssp4_pop_reg.rename(columns={'value': 'Pop(thousands)'})
    ssp4_pop_reg_1.reset_index(inplace = True)
    data4 = pd.concat([ssp4_mun_reg, ssp4_pop_reg_1["Pop(thousands)"]], axis = 1)
    data4["consum by pop (km3/thousand person)"] = data4["value"]/data4["Pop(thousands)"]

    ssp5=pd.read_csv("wat_consum_by_region_ssp5.csv")
    ssp5_mun = ssp5[ssp5["input"] == "water_td_muni_C"]
    ssp5_mun_reg = ssp5_mun[ssp1_mun["region"] == region]
    ssp5_mun_reg.reset_index(inplace = True)
    ssp5_pop = pd.read_csv("pop_by_region_ssp5.csv")
    ssp5_pop_reg = ssp5_pop[ssp5_pop["region"] == region]
    ssp5_pop_reg_1 = ssp5_pop_reg.rename(columns={'value': 'Pop(thousands)'})
    ssp5_pop_reg_1.reset_index(inplace = True)
    data5 = pd.concat([ssp5_mun_reg, ssp5_pop_reg_1["Pop(thousands)"]], axis = 1)
    data5["consum by pop (km3/thousand person)"] = data5["value"]/data5["Pop(thousands)"]

    sns.lineplot(data = data1, x = "year", y ="consum by pop (km3/thousand person)", label = "SSP1 - Sustainability")
    sns.lineplot(data = data2, x = "year", y ="consum by pop (km3/thousand person)", label = "SSP2 - Middle of the Road")
    sns.lineplot(data = data3, x = "year", y ="consum by pop (km3/thousand person)", label = "SSP3 - Regional Rivalry")
    sns.lineplot(data = data4, x = "year", y ="consum by pop (km3/thousand person)", label = "SSP4 - Inequality")
    sns.lineplot(data = data5, x = "year", y ="consum by pop (km3/thousand person)", label = "SSP5 - Fossil Fuels")
    plt.title("Municipal Water Consumption/Pop across SSPs " + region)
    plt.xlabel("Year")
    plt.ylabel("Water Consumption (km3/thousand people")
    plt.legend()


make_consum_by_pop("Africa_Northern")
```
